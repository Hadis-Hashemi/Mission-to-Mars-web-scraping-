### NASA Mars News

* Scrape the [NASA Mars News Site](https://mars.nasa.gov/news/) and collect the latest News Title and Paragraph Text. Assign the text to variables that you can reference later.

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/?page=1&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
browser.visit(url)
```

```python
for x in range(1):

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    titles = soup.find_all('div', class_='bottom_gradient')
    news = soup.find_all('div', class_='article_teaser_body')
for x in titles:
    for y in news:
        print("news_title = ", x.text)
        print("news_p = ", y.text)
        print("---------------------------------------------------------------------------------")
        
# Close the browser after scraping
    browser.quit()     
```

```python
# In this block I found a title of each image 

url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
browser.visit(url)

title_mars =[]
for x in range(1):

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    titles = soup.find_all('div', class_="description")
    for x in titles:
        title_mars.append(x.h3.text)
        

    
#title.h3.text
print(title_mars)
```
